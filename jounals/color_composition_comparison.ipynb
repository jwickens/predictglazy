```python
%matplotlib inline
import math
import matplotlib.pyplot as plt
from datasets import GlazeColor2CompositionDataset
from sklearn.decomposition import PCA
ds = GlazeColor2CompositionDataset()

pca1 = PCA(n_components=1)
x = [d for d,_ in ds]
y = [d.numpy() for _,d in ds]

x_flat = pca.fit_transform(x)
y_flat = pca.fit_transform(y)

plt.scatter(x_flat, y_flat)
plt.show()
```

```python
pca2 = PCA(n_components=2)
primary_color = [[float(i/255) for i in c[0:3]] for c in x]
secondary_color = [[float(i/255) for i in c[3:6]] for c in x]

colors = pca2.fit_transform(x)

plt.scatter([c for c,_ in colors], [c for _,c in colors], c=primary_color)
plt.show()
plt.scatter([c for c,_ in colors], [c for _,c in colors], c=secondary_color)

plt.show()
```

```python
recipes = pca2.fit_transform(y)
def scale (x):
    x = x**2
    x = math.sqrt(x)
    x = math.log(x)
    return x
plt.scatter([scale(c) for c,_ in recipes], [scale(c) for _,c in recipes], c=primary_color)
plt.show()
```

```python
from scipy.interpolate import griddata

grided_comps = griddata(recipes)

plt.scatter([c for c,_ in grided_comps], [c for _,c in grided_comps], c=primary_color)
plt.show()
```
